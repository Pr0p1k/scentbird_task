```python
from sqlalchemy import create_engine
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
engine = create_engine(f"postgresql://{os.environ['POSTGRES_USER']}:{os.environ['POSTGRES_PASSWORD']}@{os.environ['POSTGRES_HOST']}:5432/postgres")

conn = engine.connect().execution_options(stream_results=True)
```

```python
monthly_revenue = pd.read_sql("SELECT * FROM public_aprokopev.monthly_revenue", conn, chunksize=1000)
```

```python
sub = plt.subplots()
dataframe = next(monthly_revenue)
plt.plot(dataframe["report_date"].to_list(), dataframe["total_amount"].to_list())
sub[1].set(title="Monthly Revenue",xlabel="Month/year", ylabel="Amount")
plt.xticks(rotation=90)
plt.show()
```
